```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import cv2 as cv

%matplotlib inline
```

```python
df = pd.read_csv('frames_label.csv')
df = df.fillna(0)

video_source = []
frame_count = []
for file in df['file']:
    video_source.append(file.split('/')[2])
    frame_count.append(file.split('/')[-1][10:-4])

df['video_source'] = video_source
df['frame_count'] = frame_count

df
```

```python
cough = df[df['label'] == 1]
not_cough = df[df['label'] == 0]
```

```python
os.makedirs('./videos_final')
```

```python
os.makedirs('./videos_final/cough')
os.makedirs('./videos_final/not_cough')
```

```python
import time

vids = df['video_source'].unique()

for vid in vids:
    print('cough -> ', vid)
    frames = os.listdir('./videos_labelled/cough/' + vid)
    fname = './videos_final/cough/' + vid + '.avi'
    frames.sort()
    
    src = cv.imread('./videos_labelled/cough/' + vid + '/' + frames[0])
    height = np.size(src, 0)
    width = np.size(src, 1)
    
    fourcc = cv.VideoWriter_fourcc(*'MJPG')
    out = cv.VideoWriter(fname, fourcc, 8.0, (width,height))
    
    for frame in frames:
        loc = './videos_labelled/cough/' + vid + '/' + frame
        im = cv.imread(loc)
        out.write(im)
        
for vid in vids:
    print('not_cough -> ', vid)
    frames = os.listdir('./videos_labelled/not_cough/' + vid)
    fname = './videos_final/not_cough/' + vid + '.avi'
    frames.sort()
    
    src = cv.imread('./videos_labelled/not_cough/' + vid + '/' + frames[0])
    height = np.size(src, 0)
    width = np.size(src, 1)
    
    fourcc = cv.VideoWriter_fourcc(*'MJPG')
    out = cv.VideoWriter(fname, fourcc, 8.0, (width,height))
    
    for frame in frames:
        loc = './videos_labelled/not_cough/' + vid + '/' + frame
        im = cv.imread(loc)
        out.write(im)
```

```python
vids = df['video_source'].unique()

len_cough_frames = []
len_not_cough_frames = []
for vid in vids:
    frames = os.listdir('./videos_labelled/cough/' + vid)
    fname = './videos_final/cough/' + vid + '.avi'
    frames.sort()
    len_cough_frames.append(len(frames))
    print('cough ->', vid, '->' , str(len(frames)))    
        
for vid in vids:
    frames = os.listdir('./videos_labelled/not_cough/' + vid)
    fname = './videos_final/not_cough/' + vid + '.avi'
    frames.sort()
    len_not_cough_frames.append(len(frames))
    print('not cough ->', vid, '->' , str(len(frames)))    
```

```python
stat_df = pd.DataFrame(columns=['cough', 'not_cough'])
stat_df['cough'] = len_cough_frames
stat_df['not_cough'] = len_not_cough_frames
```
