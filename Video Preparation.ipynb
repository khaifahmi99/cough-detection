```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import cv2 as cv

%matplotlib inline
```

```python
df = pd.read_csv('frames_label.csv')
df = df.fillna(0)
df
```

```python
video_source = []
frame_count = []
for file in df['file']:
    video_source.append(file.split('/')[2])
    frame_count.append(file.split('/')[-1][10:-4])

df['video_source'] = video_source
df['frame_count'] = frame_count

df
```

```python
cough = df[df['label'] == 1]
not_cough = df[df['label'] == 0]
```

```python
os.mkdir('videos_labelled')
os.mkdir('./videos_labelled/cough')
os.mkdir('./videos_labelled/not_cough')
for vid_source in df['video_source'].unique():
    os.mkdir('./videos_labelled/cough/' + vid_source)
    os.mkdir('./videos_labelled/not_cough/' + vid_source)
```

```python
for index, row in cough.iterrows():
    im = cv.imread(row['file'])
    fname = './videos_labelled/cough/' + row['video_source'] + '/' + row['video_source'] + '_' + row['frame_count'] + '.jpg'
    cv.imwrite(fname, im)
    
for index, row in not_cough.iterrows():
    im = cv.imread(row['file'])
    fname = './videos_labelled/not_cough/' + row['video_source'] + '/' + row['video_source'] + '_' + row['frame_count'] + '.jpg'
    cv.imwrite(fname, im)
```
